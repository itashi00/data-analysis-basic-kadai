```python
import requests
from bs4 import BeautifulSoup
import re

# URL
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
res = requests.get(url)
res.encoding = res.apparent_encoding  # 

soup = BeautifulSoup(res.text, "html.parser")

#  <div class="main_text"> 
# <p>
paragraphs = soup.find_all("p")
text = "\n".join(p.get_text() for p in paragraphs)

# : 
text = re.sub(r'\r\n|\r|\n', '', text)
text = re.sub(r'\s+', ' ', text)
text = text.strip()

# 1
first_sentence_match = re.search(r'^(.*?)', text)
first_sentence = first_sentence_match.group(1) if first_sentence_match else text

print("1")
print(first_sentence)

# 
# 
stop_words = ['','','','','','','','','','','','','','',' ', '']

# 
# stop_words
words = re.findall(r'[\w---]+', first_sentence)
filtered_words = [w for w in words if w not in stop_words]

print("1")
print(filtered_words)
```
