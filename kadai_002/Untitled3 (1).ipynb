```python
# 📘 課題：『文壇の趨勢』の自然言語処理プログラム

# ✅ 使用ライブラリのインストールとインポート
!pip install beautifulsoup4 requests

import requests
from bs4 import BeautifulSoup
import re

# 🧩 Step1. スクレイピング：青空文庫から本文を取得
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
response = requests.get(url)
response.encoding = response.apparent_encoding

soup = BeautifulSoup(response.text, "html.parser")
main_text = soup.find("div", class_="main_text").get_text()

print("▼スクレイピング結果（先頭500文字）")
print(main_text[:500])

# 🧹 Step2. 前処理：HTMLタグ・不要な文字列の除去
cleaned_text = re.sub(r"\r|\n|\u3000", "", main_text)
cleaned_text = re.sub(r"《.*?》", "", cleaned_text)
cleaned_text = re.sub(r"［＃.*?］", "", cleaned_text)

print("\n▼前処理後のテキスト（先頭500文字）")
print(cleaned_text[:500])

# ✂️ Step3. 最初の1文のストップワード除去と品詞分解（手動）
# 対象文：
# 「近頃は大分方々の雑誌から談話をしろしろと責められて、頭ががらん空になったから、当分品切れの看板でも懸けたいくらいに思っています。」

# 手動で品詞分解し、名詞・動詞・副詞のみ抽出（ストップワード除去）
filtered_words = [
    "近頃", "大分", "方々", "雑誌", "談話", "しろしろ", "責められ", "頭",
    "がらん空", "なった", "当分", "品切れ", "看板", "懸けたい", "くらい", "思って"
]

print("\n▼ストップワード除去後の語（名詞・動詞・副詞）")
print(filtered_words)
```
